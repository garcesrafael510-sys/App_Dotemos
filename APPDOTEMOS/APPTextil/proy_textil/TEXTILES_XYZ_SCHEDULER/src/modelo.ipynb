## Programación con Restricción de Fuerza Laboral

### Conjuntos e Índices
Estos son los conjuntosn del problema:

$z \in Z$: Conjunto de Órdenes de Producción.

$i \in I_z$ : Conjunto de Productos dentro de la orden z.

$j \in J_i$ : Conjunto de Operaciones para el producto i.

$k \in K_j$: Conjunto de Actividades para la operación j.

$a \in A& : Conjunto de todas las tareas individuales a programar. 

Una tarea a es una instancia única de una actividad para una orden específica 

- Por ejemplo: la actividad *"Cerrar Lados"* para 150 unidades de la orden $O_1$
​

$m \in M$: Conjunto de tipos de máquinas  por ejemplo 'FILETEADORA', 'MANUAL'.

$t \in T$: Horizonte de planificación, discretizado en períodos de tiempo (ej. horas).

## Parámetros (Datos Conocidos)

Estos son los **"ingredientes"** que obtenemos de tus bases de datos y del entorno de producción:

- p_a : Tiempo de procesamiento (duración) de la tarea $a$. Lo calculamos a partir de los datos cargados:
$$p_a =(SAM-MINUTOS_k × CANTIDAD_REQUERIDA_z)/60 (para convertir a horas).$$

- pers_a : Personal requerido para la tarea a.

- H_t : Plantilla total disponible (número de personas) en el período de tiempo t.

- M_a: Tipo de máquina que requiere la tarea a.

- Pred(a): El conjunto de tareas predecesoras inmediatas de la tarea a. Esto lo define la jerarquía Orden -> Producto -> Operación -> Actividad y el campo CONSECUTIVO.

# Variables de Decisión

Estas son las preguntas que el modelo debe responder por nosotros:

$S_a \ge 0$: Tiempo de inicio de la tarea a.

$C_a \ge 0$: Tiempo de finalización de la tarea a.

$C_{max} \ge 0$: Makespan, el tiempo en que se completa la última tarea de todas las órdenes.

# funcion objetivo

Tu meta es maximizar el uso de los recursos, lo cual, como bien indicas, es equivalente a terminar todo el trabajo lo antes posible. Por lo tanto, el objetivo es:

Minimizar C_{max}
 
Al minimizar el tiempo total, estamos implícitamente minimizando el tiempo ocioso de los recursos (humanos y máquinas) dentro de ese lapso.

# Restricciones

1. Definición del Makespan: El makespan debe ser mayor o igual que el tiempo de finalización de cualquier tarea.

$$
C_{max} \ge C_a  \forall \ a \  \in \ A
$$

2. Duración de las Tareas: El tiempo de finalización de una tarea es su inicio más su duración.

$$
C_a = S_a + p_a \forall a \ \in A 
$$

3. Restricciones de Precedencia (Flujo de Trabajo): Una tarea no puede empezar hasta que todas sus predecesoras hayan terminado.

$$
S_a \ge C_{pred} \forall a \in A,  \forall  pred \in Pred(a)
​$$

Por ejemplo, si la actividad A_{k+1} sigue a A_k en la misma operación, entonces S_{k+1} ≥ C_k

Si la operación O_{j+1} sigue a O_j, la primera actividad de O_{j+1} no puede empezar hasta que la última actividad de O_j haya terminado.

4. Restricción de Recursos Humanos:
Esta es la restricción más compleja. Dice que, en cualquier momento $t$, la suma del personal asignado a las tareas que están activas no puede superar la plantilla disponible.

Para modelar esto formalmente, necesitamos una variable auxiliar:

- $y_{at} ∈ {0,1}:$ Es 1 si la tarea a está activa en el tiempo t, y 0 si no.

la restriccion seria

 $$\sum_{a\in A} pers_a - y_at \le H_t  \forall t  \in T$$

ahora y_{at} hay que vincularlas con las variables de inicio S_a

$$y_{at} = 1 \si solo si S_a \le t \le S_a + p_a$$
